# 1. Introduction<a name="introduction"></a>


## 1.1 Problem Setup <a name="problemsetup"></a>
- __N__ different types of molecules or chemical species 
- __M__ types of chemical reactions

Goal: to track the number of molecules of each type as a function of reaction time

Assumption: a well-mixed system, where moleculses of each type are spread uniformaly throughout the spatial domain

Suppose that initially, at time t = 0, we know how many molecules of each species are present and our aim is to describe how these numbers evolve as time increases. We therefore think of a state vector of size __N__,
$$ X(t) =  \begin{bmatrix} X_{1}(t) \\ X_{2}(t) \\ ...\\ X_{N}(t) \end{bmatrix} \quad $$

where $X_{i}(t)$ is a nonnegative integer that records how many molecules of species i are present at time t.


Our goal again is to __track state vector $X$ as a function of t__.

## 1.2 General Approaches  <a name="generalapproaches"></a>  

- __Kinetic Monte Carlo (kMC)__ , solving chemical matser equations through 
    - Gillespie's algorithm
    - Tau-leaping methods
- Reaction Rate equations through the __mean-field ODEs__

# 2. Chemical Master Equations (CME)  <a name="cme"></a>

Definition: A set of linear, autonomous ODEs. One ODE for each possible state of the system. Solution of the kth equation at time t is a real number giving the probability of system being in that particular state at time t.

State vector: $ {t,X(t)}$

The $j$th type of reaction has an associated stoichiometric, or state-change vector, $ν_{j}$ ($1 ≤ j ≤ M$), whose $i$th component is the change in the number of $S_{i}$ molecules caused by the $j$th reaction. So one reaction of type $j$ has the effect of changing the state vector from $X(t)$ to $X(t)+ ν_{j}$.

__Propensity__, $a_{j}(X(t))$, which represents the probability of this reaction taking place in the infinitesimal time interval $[t, t+dt)$. It can be written as the product of reaction rate constants and species popultaion. 

For the first order reactions, ${S_{m} \xrightarrow{\text{cj}}}$ something, $a_{j}(X(t)) = c_{j}X_{m}(t)$

For the second order reactions, ${S_{m} + S_{n} \xrightarrow{\text{cj}}}$ something, $a_{j}(X(t)) = c_{j}X_{m}(t)X_{n}(t)$

For dimerization reactions, ${S_{m} + S_{m} \xrightarrow{\text{cj}}}$ something, $a_{j}(X(t)) = c_{j}X_{m}(t)(X_{m}(t)-1)$

The chemical master equation can be written as,

$ \frac{dP(x, t)}{dt}  = \sum_{j=1}^{M} (a_{j}(x-v_{j})P(x-v_{j}, t) - a_{j}(x)P(x,t)) $

Where $X(t) = x$. $P(x,t)$ is the probability of the system being in a possible state $x$.

The dimension of the ODE is not given by the number of species, N, but by the number of possible states of the system. The latter quantity depends upon the total number of molecules present and the precise form of the chemical reactions, and is usually very large. The chemical master equations are usually hard to solve analytically since it requires enumerating the state space and computing matrix exponentials, which is infeasible for most practical systems. Hence a stochastic simulation algorithm such as Gillespie's algorithm is used here. 

# 3. Stochastic Simulation Algorithm (SSA) -Gillespie's Algorithm <a name="ssa"></a>

The kMC simulations get around the issue by computing single realizations of the state vector rather than an entire probability distribution. The problem is formulated as a __Continuous-time Markov Chain (CTMC)__ among numerous possible states. At each point in time, there is a finite probability that the system is in any given state. In practice, the PDF is estimated through statistical sampling. 

We will skip the derivation but the reader should know that two random variables are needed in each iteraction:

- Next reaction index, j
- Time until next reaction, Dt

Two random numbers ${\zeta}$ and ${z_{2}}$ in a uniform (0,1) interval are generated per iteration.

The algorithm is summarized in the flow chart below:

![Gillespie_flowchart](input_graphics/Gillespie_flowchart.PNG)

The the jth reaction is selected to fire where j is the smallest integer satisfying ${\sum_{k=1}^{j} a_{k}(X(t)) > \zeta a_{0}(X(t))}$. 

The next reaction index, j can be determined by one line of code: 

&nbsp;&nbsp;&nbsp;&nbsp; __MATLAB__ j = min(find(rand<cumsum(a/asum)))

&nbsp;&nbsp;&nbsp;&nbsp; __Python__ j = np.min(np.nonzero(np.random.rand()<np.cumsum(a/asum)))

For example, random selection of a reaction from a reaction network of 3 reactions with transition probabilities
is shown below. The arrow corresponds to a uniform random number ${\zeta}$ in (0,1). In this case, reaction 2 is selected.

# 4. Tau-leaping<a name="tau"></a>

To save the compuational cost, we consider a fixed "small enough" time interval length $\tau$ to fire simultaneously all reactions that
would have taken place. More precisely, given $X(t)$, we could freeze the propensity functions $a_{j}(X(t))$ and, using these values, fire an appropriate number of reactions of each type. This gives the tau-leaping method:

$X(t+\tau) = X(t) + \sum_{j=1}^{M} v_{j}P_{j}(a_{j}(X(t)), \tau) $

Where $P_{j}(a_{j}(X(t)), \tau)$ is a Poisson random variable with a mean and variance both equal to $a_{j}(X(t))$. A standard result from probability theory says that a Poisson random variable with large mean is well approximated by a normal random variable with the same mean and variance. Hence, in the tau-leaping method it is reasonable to change from Poisson to normal if every reaction is expected to fire many times over [t, t+$\tau$].


Replacing each $P_{j}(a_{j}(X(t)), \tau)$ with $a_{j}(X(t))\tau + \sqrt{a_{j}(X(t))\tau)}Z_{j}$ where $Z_{j}$ is an independent normal (0,1) random variables, we arrive at the __Chemical Langevin Equation (CLE)__:

$Y(t+\tau) = Y(t) + \tau \sum_{j=1}^{M} v_{j}a_{j}(Y(t)) +   \sqrt{\tau} \sum_{j=1}^{M} v_{j} \sqrt{a_{j}(Y(t))} Z_{j}$

Note: this method works well with systems having a large number of molecules.

# 5. Reaction Rate Equations (Mean-field)<a name="rre"></a>

Definitation: A set of nonlinear, autonomous ODEs. One ODE for each chemical species. Solution of the jth equation at time t is a real number representing the concentration of species j at time t.

State vector: $ {t,y(t)}$

This is a set of N ODEs involving a state vector that we write as $y(t)$. Here, $y_i(t)$ is a real number representing the concentration of the ith species at time t. 

The set of ODEs can be written as 

$ \frac{dy(x, t)}{dt}  = \sum_{j=1}^{M} v_{j}a_{j}(y(t)) $

Generally, the concentration-based rates can be written as the following:

For the first order reactions, ${S_{m} \xrightarrow{\text{kj}}}$ something, $r_{j}(y(t)) = k_{j}y_{m}(t)$

For the second order reactions, ${S_{m} + S_{n} \xrightarrow{\text{kj}}}$ something, $r_{j}(y(t)) = k_{j}y_{m}(t)y_{n}(t)$

For dimerization reactions, ${S_{m} + S_{m} \xrightarrow{\text{kj}}}$ something, $r_{j}(y(t)) = k_{j}y_{m}(t)^{2}$


# 6. Mappings Between Continuum and Miscroscopic Rate Constants<a name="mapping"></a>

By converting the concentration-based rate to a molecular rate and equating that to the propensity function, we can get a mapping between two sets of rate constants.

For the first order reactions, ${S_{m} \xrightarrow{\text{kj}}}$ something, $c_{j} = k_{j}$

For the second order reactions, ${S_{m} + S_{n} \xrightarrow{\text{kj}}}$ something, $c_{j} = k_{j}/(N_{av} vol) $

For the nth order reactions, ${S_{1} + ... + S_{n} \xrightarrow{\text{kj}}}$ something,  $c_{j} = k_{j}/(N_{av} vol)^{n-1} $

For dimerization reactions, ${S_{m} + S_{m} \xrightarrow{\text{kj}}}$ something, $c_{j} = 2k_{j}/(N_{av} vol) $

Where $N_{av}$ is avogadro's number and $vol$ is the system volume

# 7. Implementation in Python<a name="implementation"></a>



## 7.1 Case Study (a Michaelis-Menten mechanism)<a name="case"></a>

The system involves 4 species:
- a substrate, $S_{1}$
- an enzyme, $S_{2}$
- a complex, $S_{3}$
- a product, $S_{4}$

The reactions can be written as
1. $S_{1} + S_{2} \xrightarrow{\text{c1}} S_{3}$
2. $S_{3} \xrightarrow{\text{c2}} S_{1} + S_{2}$
3. $ S_{3} \xrightarrow{\text{c3}} S_{4} + S_{2}$

Where c1, c2 and c3 are rate constants.

The stoichiometeric matrix (N*M) can be written as with row i represents __species i__ and column j represents __reaction j__

$$
V = \begin{bmatrix}
-1&1&0\\
-1&1&1\\
1&-1&-1\\
0&0&1\\
\end{bmatrix}
$$


```python
# Imports necessary libraries
import time
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import odeint, ode

# Set random seed
np.random.seed(0)
```

## 7.2 Gillespie's Algorithm<a name="gillespie"></a>

```python
# Stoichiometric matrix
V = np.array([(-1, 1, 0), (-1, 1, 1), (1, -1, -1), (0, 0, 1)])

# Constants and Initial Conditions
nA = 6.023e23 #Avagadro's number
vol = 1e-15 #volume of system
```

```python
# Molecule counts
X = np.zeros(4)
X[0] = round(5e-7*nA*vol) # molecules of substrate
X[1] = round(2e-7*nA*vol) # molecules of enzyme

# Reaction constants
c=np.zeros((3,1))
c[0] = 1e6/(nA*vol) #c1
c[1] = 1e-4 #c2
c[2] = 0.1 #c3

# Time values
t = 0
tfinal = 50
```

```python
# Create an array to store time, # of substrate and # of products
data_ga = np.array([(t, X[0], X[3])])

start = time.time()
print('Starting kMC simulations...')

# Run the simulation
# Updates concentrations over timeframes until tfinal is reached
while t < tfinal:
    
    # Determines likelyhood of each reaction to occur
    a = np.zeros((3,1))
    a[0] = c[0]*X[0]*X[1]
    a[1] = c[1]*X[2]
    a[2] = c[2]*X[2]
    asum = np.sum(a)
    # Uses random number to determine which reaction occurs
    j = np.min(np.nonzero(np.random.rand()<np.cumsum(a/asum)))
    # Uses random number to determine time for reaction
    tau = np.log(1/np.random.rand())/asum
    # Updates concentrations and time
    X = X + V[:,j]
    t = t + tau
    
    data_ga = np.append(data_ga, [(t, X[0], X[3])], axis = 0)
    
end = time.time()
print('The simulation takes {0:.4f} s'.format(end-start))
```

### Plot the numbers of substract, product vs time

```python
fig_ga = plt.figure()
ax_ga = fig_ga.add_subplot(111)
ax_ga.step(data_ga[:,0],data_ga[:,1],'g-o', markerfacecolor = 'g',
    markeredgecolor = 'g')
ax_ga.step(data_ga[:,0],data_ga[:,2],'r-*', markerfacecolor = 'r',
    markeredgecolor = 'r')
ax_ga.set_title('GILLESPIE\'S ALGORITHM')
ax_ga.set_xlabel('Time')
ax_ga.set_ylabel('Molecules');
ax_ga.text(40,240,'Product');
ax_ga.text(30,40,'Substrate'); 
```

### Plot a section of graph (10-20 s) to show detail in molecule change

```python
fig_ga_zoom = plt.figure()
ax_ga_zoom = fig_ga_zoom.add_subplot(111)
ax_ga_zoom.step(data_ga[:,0],data_ga[:,1],'g-o', markerfacecolor = 'g',
    markeredgecolor = 'g')
ax_ga_zoom.step(data_ga[:,0],data_ga[:,2],'r-*', markerfacecolor = 'r',
    markeredgecolor = 'r')
ax_ga_zoom.set_title('GILLESPIE\'S ALGORITHM')
ax_ga_zoom.set_xlabel('Time')
ax_ga_zoom.set_ylabel('Molecules');
ax_ga_zoom.set_xlim(10,20);
ax_ga_zoom.set_ylim(95,120);
```

## 7.3 Tau-leaping<a name="tauleaping"></a>

```python
# Stoichiometric matrix
V = np.array([(-1, 1, 0), (-1, 1, 1), (1, -1, -1), (0, 0, 1)])

# Constants and Initial Conditions
nA = 6.023e23
vol = 1e-15
# Molecule counts
Y = np.zeros(4)
Y[0] = round(5e-7*nA*vol)
Y[1] = round(2e-7*nA*vol)
# Reaction constants
c=np.zeros((3,1))
c[0] = 1e6/(nA*vol);
c[1] = 1e-4; 
c[2] = 0.1;
# Time values
tfinal = 50
L = 250

# Define a "small enough tau"
tau = tfinal/L
```

```python
# Create an array to store time, # of substrate and # of products
data_ts = np.array([(0, Y[0], Y[3])])

start = time.time()
print('Starting kMC simulations...')

# Run the simulations
# Updates concentrations over timeframes until tfinal is reached
for k in range(L):
    # Probabilities of reactions
    a=np.zeros((3,1))
    a[0] = c[0]*Y[0]*Y[1];
    a[1] = c[1]*Y[2];
    a[2] = c[2]*Y[2];
    # Uses normally distributed random numbers to determine extents if reaction
    d=np.zeros((3,1))
    d[0] = tau*a[0] + np.sqrt(abs(tau*a[0]))*np.random.randn()
    d[1] = tau*a[1] + np.sqrt(abs(tau*a[1]))*np.random.randn()
    d[2] = tau*a[2] + np.sqrt(abs(tau*a[2]))*np.random.randn()
    # Updates concentrations
    Y = Y + d[0]*V[:,0] + d[1]*V[:,1] + d[2]*V[:,2]
    data_ts = np.append(data_ts, [(k*tau, Y[0], Y[3])], axis = 0)
    
end = time.time()
print('The simulation takes {0:.4f} s'.format(end-start))
```

### Plot the numbers of substract, product vs time

```python
# Plots full data set
fig_ts = plt.figure()
ax_ts = fig_ts.add_subplot(111)
ax_ts.plot(data_ts[:,0],data_ts[:,1],'g-o', markerfacecolor = 'g',
    markeredgecolor = 'g')
ax_ts.plot(data_ts[:,0],data_ts[:,2],'r-*', markerfacecolor = 'r',
    markeredgecolor = 'r')
ax_ts.set_title('TAU-LEAPING')
ax_ts.set_xlabel('Time')
ax_ts.set_ylabel('Molecules')
ax_ts.text(40,240,'Product')
ax_ts.text(30,50,'Substrate');
```

## 7.4 Mean-Field ODEs<a name="ode"></a>

```python
# Sets timeframe and initial conditions
tspan = np.array([0,50])
y0 = np.array([5e-7, 2e-7, 0, 0])
# Reaction constants
k1 = 1e6
k2 = 1e-4
k3 = 0.1
# Array to store data
data_ode = []
```

```python
def kin(t,y):
    '''
    Uses the kinetics of each reaction based on concentrations to model changes
    in the concentrations
    '''
    yprime = np.zeros(4);
    yprime[0] = -k1*y[0]*y[1] + k2*y[2];
    yprime[1] = -k1*y[0]*y[1] + (k2+k3)*y[2];
    yprime[2] = k1*y[0]*y[1] - (k2+k3)*y[2];
    yprime[3] = k3*y[2];
    return yprime
```

```python
def solout(t,y):
    # Stores data as reaction is modeled
    data_ode.append([t,*y])
```

```python
# Constructs ode solver
y=ode(kin).set_integrator('dopri5',rtol=1e-8,method='bdf')
y.set_solout(solout)
# Provides initial conditions and solves the ode
y.set_initial_value(y0,tspan[0])

start = time.time()

# Solve the odes
y.integrate(tspan[1])

end = time.time()
print('Solving odes takes {0:.4f} s'.format(end-start))

data_ode = np.array(data_ode)
```

```python
fig_ode = plt.figure()
ax_ode = fig_ode.add_subplot(111)
ax_ode.plot(data_ode[:,0],data_ode[:,1],'g-o', markerfacecolor = 'g',
         markeredgecolor = 'g')
ax_ode.plot(data_ode[:,0],data_ode[:,4],'r-*', markerfacecolor = 'r',
         markeredgecolor = 'r')
ax_ode.set_title('MEAN-FIELD ODEs')
ax_ode.set_xlabel('Time')
ax_ode.set_ylabel('Concentration')
ax_ode.text(33,3.5e-7,'Product')
ax_ode.text(33,0.5e-7,'Substrate');
```
